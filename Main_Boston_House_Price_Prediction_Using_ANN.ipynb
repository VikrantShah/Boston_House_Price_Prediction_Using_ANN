# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the dataset

```python
column_names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
```

```python
dataset = pd.read_csv(r"Dataset/housing.data", header = None, delimiter = r"\s+", names = column_names)
```

```python
dataset.head()
```

```python
X = dataset.drop(columns = ["MEDV"])
y = dataset[["MEDV"]]
```

```python
print(X)
```

```python
print(y)
```

# Splitting the dataset into Training and Testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

# ANN Implementation

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
```

```python
regressor = Sequential()
```

## Adding the layers

```python
# Adding the input and first hidden layer
regressor.add(Dense(units = 8, activation = "relu", input_shape = (13, )))

# Adding the second hidden layer
regressor.add(Dense(units = 16, activation = "relu"))

# Adding the third hidden layer
regressor.add(Dense(units = 32, activation = "relu"))

# Adding the forth hidden layer
regressor.add(Dense(units = 32, activation = "relu"))

# Adding the output layer
regressor.add(Dense(units = 1))
```

## Compiling the ANN

```python
regressor.compile(optimizer = "adam", loss = "mean_squared_error", metrics = ["accuracy"])
```

## Training the ANN

```python
regressor.fit(X_train, y_train, batch_size = 32, epochs = 1750, callbacks = [EarlyStopping(patience = 7)])
```

```python
regressor.summary()
```

```python
regressor.save("model.h5")
```

# Predicting the results of the Testing set

```python
y_pred = regressor.predict(X_test)
```

```python
np.set_printoptions(precision = 2)
# Set the number to output after decimal

print(np.concatenate((np.array(y_pred).reshape(len(y_pred), 1), np.array(y_test).reshape(len(y_test), 1)), 1))
# 1 : Converts it from horizontal to vertical array
```

# Plotting the predicted and actual results of Testing set

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual MEDV values")
plt.ylabel("Predicted MEDV values")
plt.title("Boston House Price Prediction")
plt.show()
plt.savefig("Predicted_VS_Actual.jpg")
```

# Calculating the Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
print(mse)
```
